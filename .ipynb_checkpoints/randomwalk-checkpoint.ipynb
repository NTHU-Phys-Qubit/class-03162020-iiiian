# importing mod

```python
import numpy as np
import matplotlib.pyplot as plt
```

# a 1d random walk with step size between -0.5 and 0.5

```python
n_steps = np.arange(1001)
position = np.array([0])
ran = np.random.random((1000,))
ran = ran - 0.5

for i in range(1000):
    position = np.append(position, position[i] + ran[i])

plt.plot(n_steps, position)   
plt.xlabel("n steps")
plt.ylabel("position")
```

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in 1d array
"""

def ran_walk_continue(steps, n):
    ans = np.zeros((n,))
    for i in range(n):
        ran = np.random.random((steps,))
        ran = ran - 0.5
        final_place = np.sum(ran)
        ans[i] = final_place
    return ans
```

```python
ans1 = ran_walk_continue(1000, 10000)
plt.hist(ans1, bins=50)
plt.show()
```

# a 1d random walk with discrete step size (-0.5 or 0.5)

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in 1d array
"""

def ran_walk_discrete(steps, n):
    ans = np.zeros((n,))
    for i in range(n):
        ran = np.random.randint(2, size=steps)
        ran = ran - 0.5
        final_place = np.sum(ran)
        ans[i] = final_place
    return ans
```

```python
ans2 = ran_walk_discrete(1000, 10000)
plt.hist(ans1, bins=50)
plt.show()
```

# a 2d random walk with fixed step size 1 but the direction is random

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in array of shape (1,2), which is x and y position repectively
"""

def rand_angle_2d(steps, n):
    ans_x = np.zeros((n,))
    ans_y = np.zeros((n,))

    for i in range(n):
        ran = np.random.random((steps,))
        ran = ran * 2 * np.pi
        x = np.sin(ran)
        y = np.cos(ran)
        ans_x[i] = np.sum(x)
        ans_y[i] = np.sum(y)

    ans = np.array([ans_x, ans_y])
    return ans
```

```python
ans3 = rand_angle_2d(1000, 10000)
plt.hist2d(ans3[0], ans3[1], bins=50)
plt.colorbar()
plt.show()
```

# a 2d random walk with a custom discrete step size on x and y axis

```python
"""
n_steps: the steps each random walk take
step_size: the size of a step on on x and y axis
n: calculate n samples of random walks

returns the final position of each random walk in array of shape (1,2), which is x and y position repectively
"""


def ran_step_2d(n_steps, step_size, n):
    ans_x = np.zeros((n,))
    ans_y = np.zeros((n,))
    for i in range(n):
        ran_x = np.random.randint(2, size=n_steps)
        ran_y = np.random.randint(2, size=n_steps)
        ran_x = 2 * step_size * (ran_x - 0.5)
        ran_y = 2 * step_size * (ran_y - 0.5)
        ans_x[i] = np.sum(ran_x)
        ans_y[i] = np.sum(ran_y)
    ans = np.array([ans_x, ans_y])
    return ans
```

```python
ans4 = ran_step_2d(1000, 5, 10000)
plt.hist2d(ans4[0], ans4[1], bins=50)
plt.colorbar()
plt.show()
```
